### CREATING ARRAY

```python
a = np.array([1,2,3,4,5,6])
print(a)
```

```python
a = np.array([[1,2,3,4,5,6],[1,2,3,4,5,6],[2,3,5,6,7,8]])
print(a)
```

```python
a = np.array([0,0,0,0])
print(a)
```

```python
zero = np.zeros((2,10))
print(zero)
```

```python
print(np.ones((2,5)))
```

```python
print(np.full((2,6),9))
```

```python
print(np.arange(10))
```

```python
print(np.eye(3)) #identity matrix
```

```python
rng = np.random.default_rng(seed=42)
print(rng.random((2,3)))
```

```python
print(rng.integers(1,10,size=(3,3)))
```

```python
print(rng.normal(loc=0, scale = 1,size = 5))
```

```python
print(M.size)
```

### INDEXING AND SLICING

```python
x = np.arange(10,20)
print(x)
```

```python
print(x[2])
```

```python
M = np.array([[1,2,3,4],[3,4,5,6]])
n = np.arange(1,13).reshape(3,4)
print(M)
print(n)
```

```python
print(M[1])
print(M[0][3])
```
