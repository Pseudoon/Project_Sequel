```python
pop = pd.Series([1444,2324,234,2342,2344],
               index = ["ok","good","v good","v v good","excellent"],
                name = "reviews"
               )
```

```python
print(pop)
```

```python
print(pop[pop > 500 ])
```

```python
raw = pd.DataFrame({
    "order_id" : ['1',"2","3"],
    "amount" : ['234',"4350","5650"]
})
```

```python
print(raw["amount"].sum())
```

```python
clean = raw.copy()
```

```python
clean["order_id"] = pd.to_numeric(clean["order_id"])
clean["amount"] = pd.to_numeric(clean["amount"])
```

```python
print(clean["amount"].sum())
```
