**Q20.** Create a column vector `[[1], [2], [3], [4]]` (shape `(4,1)`) and add it to the row vector `[10, 20, 30]` (shape `(3,)`). What is the resulting shape? Why?

````python
col_vec = np.array([[1], [2], [3], [4]])  
row_vec = np.array([10, 20, 30])        

result = col_vec + row_vec
print("Result:", result)
print("Resulting Shape:", result.shape)
````

**Q21.** Try adding a `(2, 3)` array and a `(4,)` array. Catch the `ValueError` with try/except and print the error message. Explain in a comment why broadcasting fails here.

````python
arr_a = np.ones((2, 3))
arr_b = np.ones((4,))

try:
    result = arr_a + arr_b
except ValueError:
    print(f"could not be broadcast together with shapes (2,3) (4,)")
````

## Section 6 — Aggregations & the `axis` Argument

**Q22.** Given
```python
M = np.array([[3, 7, 2, 8],
              [5, 1, 9, 4],
              [6, 0, 3, 2]])
```
Compute:
- (a) total sum of all elements
- (b) sum of each column
- (c) sum of each row
- (d) maximum of each row
- (e) minimum of each column

````python
M = np.array([[3, 7, 2, 8],
              [5, 1, 9, 4],
              [6, 0, 3, 2]])
print("Total sum :",M.sum())
print("Total sum column :",M.sum(axis=0))
print("Total sum row:",M.sum(axis = 1))
print("maximum of each row :",M.max(axis = 1))
print("minimum of each row :",M.min(axis = 0))
````

**Q23.** For the same matrix `M`, find the index of the maximum element in each column using `argmax`.

````python
max_index_col = M.argmax(axis=0)
print("Index of max element in each column:", max_index_col)
````

**Q24.** Compute the overall mean and standard deviation of `M`, rounded to 2 decimal places.

````python
mean_val = np.round(M.mean(), 2)
std_val = np.round(M.std(), 2)

print(f"Overall Mean: {mean_val}")
print(f"Overall Std Dev: {std_val}")
````

**Q27.** Generate a 1-D array of 20 random numbers drawn from a standard normal distribution (seed = 0). Then compute its mean, std, min, and max.

````python
rng = np.random.default_rng(seed=0)
rand_arr = rng.standard_normal(20)
print(rand_arr.mean())
print(rand_arr.std())
print(rand_arr.min())
print(rand_arr.max())
````

**Q28.** Normalize the columns of a 4×3 random matrix so that each column has mean 0 and standard deviation 1. (Use broadcasting — no loops.)

````python
rng = np.random.default_rng(seed=42)
X = rng.random((4, 3))

col_mean = X.mean(axis=0)
col_std = X.std(axis=0)
X_nor = (X -col_mean)/col_std
print(X_nor)
print("new mean",X_nor.mean(axis=0))
print("new std",X_nor.std(axis=0))
````

**Q29.** Given marks of 5 students in 3 subjects:
```python
marks = np.array([[78, 82, 90],
                  [65, 70, 72],
                  [88, 95, 92],
                  [54, 60, 58],
                  [80, 85, 87]])
```
Find: (a) each student's average, (b) each subject's average, (c) the index of the top-scoring student overall (highest total), (d) the subject in which the class performed best on average.

````python
marks = np.array([[78, 82, 90],
                  [65, 70, 72],
                  [88, 95, 92],
                  [54, 60, 58],
                  [80, 85, 87]])
avg = marks.mean(axis=1)
print(np.round(avg,2))
subject_avg = marks.mean(axis=0)
print(np.round(subject_avg,2))
student_totals = marks.sum(axis=1)
top_student_idx = student_totals.argmax()
print(student_totals,top_student_idx)
````

**Q30.** Create a 10×10 multiplication table (where `M[i, j] = (i+1) * (j+1)`) using broadcasting only.

````python
rows = np.arange(1, 11).reshape(10, 1) 
cols = np.arange(1, 11) 

multiplication_table = rows * cols
print(multiplication_table)
````
