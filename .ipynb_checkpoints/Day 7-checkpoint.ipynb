```python
from sklearn.impute import SimpleImputer
```

```python
num_values = df.select_dtypes(include=["number"]).columns
num_imputer = SimpleImputer(strategy="median")
df[num_values] = num_imputer.fit_transform(df[num_values])
cat_values = df.select_dtypes(include=["str"]).columns
cat_imputer = SimpleImputer(strategy="most_frequent")
df[cat_values] = cat_imputer.fit_transform(df[cat_values])
```

```python
print(df.isnull().sum())
print((df.isnull().mean()*100).round(1))
```

```python
mask = pd.Series(True, index=df.index)
for colin ['Age', 'Salary']:
    Q1, Q3 =df[col].quantile(0.25), df[col].quantile(0.75)
IQR = Q3- Q1
mask &=df[col].between(Q1- 1.5*IQR,Q3 + 1.5*IQR)
df = df[mask].reset_index(drop=True)
```

```python
edu_order ={'High School': 0, 'Bachelor': 1, 'Master': 2, 'PhD': 3}
df['Education_Label'] = df['Education'].map(edu_order)
```
