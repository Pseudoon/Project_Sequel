### 19. Mini Challenge

**Question:** Write a program that takes a list of marks `[45, 78, 89, 56, 92, 67]` and prints: the average, the highest, the lowest, and a list of marks that are above the average (use list comprehension).

```python
list_marks = [45, 78, 89, 56, 92, 67]
print("Max Marks :",max(list_marks))
print("Min Marks :",min(list_marks))
avg = sum(list_marks)/len(list_marks)
print("Average:",avg)
print("list of marks that are above the average :",[x for x in list_marks if x > avg])
```
