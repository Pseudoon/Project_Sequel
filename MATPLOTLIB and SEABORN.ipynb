## MATPLOTLIB
axes : individual plots inside the figure

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import random
```

```python
x = np.linspace(0,10,100)
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y,label= "Sin(x)",color = "red")
plt.title("Sine Wave Plot")
plt.xlabel("x")
plt.ylabel("Sin(x)")
plt.legend()
plt.show()
```

```python
fig,ax = plt.subplots()
ax.plot(x,y,label = "sin(x)",color = "black")
ax.set_xlabel("x")
ax.set_ylabel("sin(x)")
ax.set_title("Sine Wave")
ax.legend()
plt.show()
```

```python
np.random.seed(42)
epochs = np.arange(1,21)
```

```python
train_loss = np.exp(-epochs/5) + np.random.rand(20)*0.5
```

```python
val_loss = np.exp(-epochs/6) + np.random.rand(20)*0.08
```

```python
plt.plot(epochs,train_loss,label ="train")
plt.plot(epochs,val_loss,label = "validity")
plt.title("Training vs Validation Loss")
plt.xlabel("epochs")
plt.ylabel("loss")
plt.legend()
plt.show()
```

## BARPLOT

```python
models = ['logReg',"RF","SVM","XGB"]
acc = [0.82,0.45,0.67,0.93]
```

```python
plt.bar(models,acc,color="red",label = "model Accuracy")
plt.title("Accuracy of Models")
plt.ylabel("Accuracy")
plt.xlabel("Models")
plt.legend()
plt.show()
```

## HISTOGRAM

## MULTIPLOT

```python
fig, ax = plt.subplots(1,2,figsize = (10,4))
ax[0].plot(epochs, train_loss,color="yellow")
ax[1].hist(data,bins=40,color="skyblue",edgecolor="black")
```

# SEABORN

```python
iris = sns.load_dataset("iris")
tips = sns.load_dataset("tips")
```

```python
print(iris.head())
print(tips.head())
```

```python
sns.histplot(data=iris, x="sepal_length", hue="species", kde=True)
```

```python
sns.boxplot(data = iris, x = "species", y="petal_length")
plt.title("Box plot")
plt.show()
```

```python
sns.violinplot(data = iris, x="species", y="petal_width")
plt.title("Violin Plot")
```

```python
corr = iris.drop(columns=["species"]).corr()
```

```python
sns.heatmap(corr, annot =True)
```

```python
sns.scatterplot(data=iris, x = "sepal_length", y = "petal_length", hue = "species")
plt.title("Scatter with Hue")
```

```python
sns.countplot(data = tips, x="day",hue="sex")
```

```python
sns.countplot(data = iris, x="species")
```

```python
sns.pairplot(data=iris,hue="species")
```
