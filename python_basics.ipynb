## list comprehension

```python
num = [1,2,3,4,5]
square = []
for x in num:
    square.append(x ** 2)

print(square)
```

```python
sq = [i**2 for i in num]
print(sq)
```

```python
even = [x for x in range(20) if x%2 == 0]
print(even)
```

```python
w = ["python","hello","world"]
up = [i.upper() for i in w]
print(up)
```
